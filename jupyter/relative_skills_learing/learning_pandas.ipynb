```python
# 1. 直接创建一个名为 'ones' 的新列，并赋值为 1
# Pandas 会自动进行“广播”，把这一列 97 行全部填满 1
data.insert(0, 'ones', 1) 

# 2. 现在查看 data.head()，你会发现第一列变成了 ones
print(data.head())
#data.insert(位置, 列名, 数值)
```

```python
data_iloc=data.iloc[:,-1:]
#df.iloc[row_index, column_index]
# 若是:则是填入所有的数据
# 若是填入一个数字取出为一维数组，无法进行向量运算，要加上一个:(例如-1:就是取出最后一行变成一个n*1的矩阵)
# 想要取出剩一行则是:+要去掉的那一行
# 若是想要取出从哪一行到另一行则是，形似0:5这样开始加上结尾的形式
data_iloc.head()
```

```python
data.dropna(axis=0,how='any',subset=None)
# #df.dropna(axis=0, how='any', subset=None)清空不当数据
# axis: 0 表示删行，1 表示删列。
# how: 'any' 表示只要有一个空值就删，'all' 表示全为空才删。
data.head()
```

```python
data.describe()# 描述数据
#df.describe(percentiles=None, include=None)
# include: 'all' 会包含非数值列，默认只看数值列。
```
