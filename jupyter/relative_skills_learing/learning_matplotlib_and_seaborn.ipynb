```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_data=pd.read_csv("ex1data1.txt",names=["population","profit"])
my_data.head()
```

```python
loss_history = [32, 20, 10, 5, 4.5]
plt.plot(loss_history, 'b-', label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Cost J')
plt.legend()
plt.show() 
# plt.plot(x, y, format_str, label=None)

# x, y: 数据点。可以是列表、NumPy 数组或 Pandas 的 Series。
# format_str: 控制颜色和形状。如 'r--'（红色虚线）、'bo'（蓝色圆点）。
# label: 给这条线起个名字，配合 plt.legend() 显示图例。
```

```python
plt.scatter(my_data["population"],my_data["profit"],c="red",marker='x')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
# plt.scatter(x, y, s=None, c=None, marker=None)

# s: 点的大小 (size)。
# c: 颜色 (color)。
# marker: 形状（如 'x', 'o', '*'）。
```

```python
#fig, ax = plt.subplots(rows, cols, figsize=(width, height))
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
# 在第一个子图画散点
ax[0].scatter(my_data["population"],my_data["profit"],c="red",marker='x')
ax[0].set_xlabel('Population of City in 10,000s')
ax[0].set_ylabel('Profit in $10,000s')

# 在第二个子图画 Loss
ax[1].plot(loss_history, 'b-', label='Training Loss')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Cost J')
ax[1].legend()

plt.show()
```

```python
# sns.lmplot(x, y, data, line_kws=None)
# x, y: 字符串，对应 DataFrame 的列名。
# data: 传入整个 DataFrame。
#line_kws: 字典，用来改线的颜色，比如 {'color': 'red'}。

sns.lmplot(x='population', y='profit', data=my_data)
plt.show()
```
